```python
import cv2 as cv
import numpy as np
flags = [i for i in dir(cv) if "COLOR_" in i]
flags
```

```python
cap = cv.VideoCapture("../resource/worker-zone-detection.mp4")
while (1):
    _, frame = cap.read()


    '''
    green = np.uint8([[[0,255,0 ]]])
    hsv_green = cv.cvtColor(green,cv.COLOR_BGR2HSV)
    print( hsv_green )
    [[[ 60 255 255]]]

    '''

    # Convert bgr to gray
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)

    # # # Define hsv
    # lower_blue = np.array([110, 50, 50])
    # upper_blue = np.array([130, 255, 255])
    # # lower_blue = np.array([0, 0, 200])  # Low saturation and high value
    # # upper_blue = np.array([179, 20, 255])  # High range for brightness (value)


    # Define the red color range in HSV space
    # Lower range of red (around 0 to 10)
    lower_red1 = np.array([0, 120, 70])  
    upper_red1 = np.array([10, 255, 255])

    # Upper range of red (around 170 to 180)
    lower_red2 = np.array([170, 120, 70])
    upper_red2 = np.array([180, 255, 255])

    # Threshold the HSV image to get red colors
    mask1 = cv.inRange(hsv, lower_red1, upper_red1)
    mask2 = cv.inRange(hsv, lower_red2, upper_red2)

    # Combine the masks to cover both ranges of red
    mask = cv.bitwise_or(mask1, mask2)
    # Threshold hsv image to get only blue colors
    # mask = cv.inRange(hsv, lower_blue, upper_blue)

    # Bitwise-and mask and original image
    res = cv.bitwise_and(frame, frame, mask=mask)

    # cv.imshow('frame',frame)
    cv.imshow('frame', res)

    # cv.imshow('mask',mask)
    # cv.imshow('res',res)
    k = cv.waitKey(5) & 0xFF
    if k == 27:
        break
 
cv.destroyAllWindows()
```
